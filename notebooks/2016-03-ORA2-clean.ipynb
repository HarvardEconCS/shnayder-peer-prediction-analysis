```python
by_criterion = grouper_by_keys("course_id", "item_id", "criterion_name")
just_peer.sort(key=by_criterion)

# (course_id, item_id, criterion_name) -> A list of assessments
assessments_by_criterion = dict([(k, list(group))
                                 for k, group in groupby(just_peer, by_criterion)])
```

```python
def criterion_hist(assessments):
    """
    Args:
        assessments: dict: criterion -> list of assessments
    
    Return a distribution (list) of num assessments per criterion, ready for plotting.
    """
    return [len(xs) for xs in assessments.values()]


def group_by_submission(assessments):
    """
    sorts, then groups by submission. Returns groupby return value:
        iter over (k, group_iter)
    """
    by_submission = lambda x: x.submission_uuid
    assessments.sort(key=by_submission)

    return groupby(assessments, by_submission)


def submission_hist(assessments):
    """
    Args:
        assessments: dict: criterion -> list of assessments
    
    Groups assessments for each criterion by submission, return 
    a distribution (list) of num assessments per submission, ready for plotting.
    """
    
    return list(chain.from_iterable([[len(list(xs))
                                      for k, xs in group_by_submission(vs)]
                                     for vs in assessments.values()]))    
```

```python
def plot_hist(data, title="", bins=10, **kwargs):
    fig, ax = plt.subplots(**kwargs)
    ax.hist(data, bins=bins)
    ax.set_title(title)
    return fig, ax
```

```python
plot_hist(criterion_hist(assessments_by_criterion), "# Assessments / criterion")
plot_hist(submission_hist(assessments_by_criterion), "# Assessments / submission")
```
