```python
def save_figure(fig, filename):
    """
    Save fig in FIG_DIR/filename, with bbox_inches=tight
    """
    fig.tight_layout()
    fig.savefig(FIG_DIR + "/" + filename , bbox_inches='tight')
```

```python
def plot_cdf(counts, title="", ax=None, fig_kwargs=None, **kwargs):
    """
    Given a counter dict (e.g. {1:3, 4:7, 2:5}),
    plot a corresponding cdf (e.g. for 1 1 1 2 2 2 2 2 4 4 4 4 4 4 4).
    Example points: (0,0), (0,1), (3,1), (3,4), (8,4), (11,), ()
    """
    # (1,3), (2,5), (4,7)
    pairs = sorted(counts.items())
    
    # (1,2,4), (3,5,7)
    vals, widths = zip(*pairs)
    
    # 0,3,8,15
    lefts = [sum(widths[:k]) for k in range(len(widths)+1)]
    # duplicate the last element
    vals = vals + (vals[-1],)
    
    #print lefts, vals, widths

    if ax is None:
        if fig_kwargs is None:
            fig_kwargs = {}
        fig, ax = plt.subplots(**fig_kwargs)
        
    ax.step(vals, lefts, where='pre',
            alpha=plotting.DEFAULT_ALPHA)
    ax.set_title(title)
    ax.set_xlim(xmin=0)
    return ax
```

```python
def scores_by_size(assessments, scoring_fn):
    """
    Args:
        assessments -- must be sorted by criterion
        scoring_fn -- fn: assessments for criterion -> scores
    Returns:
        dict: size -> {score -> count}
    """
    score_counts = defaultdict(Counter)  # size -> {score -> count}
    for i, (k, xs) in enumerate(groupby(assessments, key=by_criterion)):
        xs = tuple(xs)
#         print '.',
#         if i%50 == 0:
#             print ''
        submissions = edx.group_scores_by_submission(xs)
        size = max(md.flatten(submissions))+1
        counts = Counter(scoring_fn(xs))
        score_counts[size] += counts
    return score_counts
```

### Truthful scores vs random

How much better is it to be truthful than random? Given bounded score range, can't scale up indefinitely. Connected to variance.

Let's start by looking at histogram of (truthful - random) by criterion, then do the same thing weighted by counts.

```python
# Let's start with 01, since we can easily compute that from the joint
jointfile = data_path + "signal_joint_20160422.json"
# Load data
with open(jointfile, 'r') as f:
    # save as an array because keys aren't strings
    lst = json.load(f)
    signal_joints = dict((tuple(k), np.array(v)) for k,v in lst)
    
# organize signal joints by size
joints_by_size = defaultdict(dict)  # size -> {criterion -> joint}
for k, v in signal_joints.items():
    joints_by_size[v.shape[0]][k] = v
```

```python
def benefit01():
    deltas = map(md.delta_from_joint, signal_joints.values())
    truthful_scores = [md.expected_01_score(d) / 2.0 + 0.5 for d in deltas]
    diffs = [t - 0.5 for t in truthful_scores]
    return diffs


fig, ax = plt.subplots()
ax.hist(benefit01(), bins=20, lw=2,#alpha=plotting.DEFAULT_ALPHA,
        cumulative=-1, normed=1, histtype='step',label="CA")
ax.set_xlabel("Score threshold")
ax.set_ylabel("Fraction of criteria")
ax.legend(loc="upper right")

sns.despine(fig)
plotting.set_width(fig, "half")
```
